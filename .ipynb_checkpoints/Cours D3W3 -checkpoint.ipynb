```python
Vehicle = pd.read_excel('/Users/dehilinesma/Downloads/vehicles/vehicles.xlsx')
Vehicle
```

```python
#Adding and 
Vehicle .loc[Vehicle ['Transmission'].str.startswith('A'), 'TransType'] = 'Automatic'
Vehicle.loc[Vehicle ["Transmission"].str.startswith('M'), 'TransType']= 'Manual'

Vehicle
```

```python
# Combine the Dataframe
avg_mpg = Vehicle .groupby('Make', as_index=False).agg({'Combined MPG':'mean'})
avg_mpg.columns = ['Make', 'Avg_MPG']

avg_mpg
```

```python
car= pd.merge(Vehicle , avg_mpg, on='Make')
car.head(10)
```

```python
sum_mpg = Vehicle .groupby('Make', as_index=False).agg({'Combined MPG':'sum'})
sum_mpg.columns = ['Make', 'sum_MPG']

sum_mpg
```

```python
car1= pd.merge(car, sum_mpg, on='Make')
car1
```

```python
#filtering
filtered = car1[(car1['Make']=='Ford') & 
                (car1['Cylinders']>=6) & 
                (car1['Combined MPG'] < 18)]


filtered.head()
```

```python
filtered['Make'].value_counts()
```

```python
mpg_labels = ['Very Low', 'Low', 'Moderate', 'High', 'Very High']

car1['bins'] = pd.cut(car1['Combined MPG'],5, labels=mpg_labels)
car1.head(10)
# cut by values 
# ('mix-min')/our labels 
```

```python
car1['bins1'] = pd.qcut(car1['Combined MPG'],5, labels=mpg_labels)
car1.head(10)
#Quartiles
```

```python
cutoffs = [7,14,21,23,30,40]
car1['bins2'] = pd.cut(car1['Combined MPG'],cutoffs, labels=mpg_labels)
car1.head(10)

#Identify the outliers and cut it off 
# higher than the minimum 
```
