```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([(1,231,"Self-employed","009","High", 2, 30200),
                   (2,765,"students","005","high",3,12700),
                   (3,453,"Horeca","007","medium",5,89400),
                   (4,231,"self-employed","009","high",2,30200),
                   (5,892,"finance","003","low",3,740000),],
                  index=["1", "2", "3", "4", "5"],
columns=("TransactionID","ClientID","Profession","Bank_dep","Risk","Number of credits","Revenue",),
)
df
```

```python
df['Profession'] = df['Profession'].str.capitalize()
df
```

```python
df['Risk'] = df['Risk'].str.capitalize()
df
```

```python
df.dtypes
```

```python
# Method 1: Dummy Encoder
a=pd.get_dummies(df, prefix='object')
print(a.head())
```

```python
# Method 2: Label Encoder 

from sklearn import preprocessing 
label = preprocessing.LabelEncoder() 
df['Profession']= label.fit_transform(df['Profession']) 
df['Bank_dep']= label.fit_transform(df['Bank_dep']) 
df['Risk']= label.fit_transform(df['Risk']) 

print(df.head())
```
