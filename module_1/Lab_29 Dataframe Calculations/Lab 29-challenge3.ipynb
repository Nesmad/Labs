We break down the main problem into several sub problems:

```python
amount_spent = df.groupby('CustomerID', as_index=False).agg({'amount_spent':'sum'}).sort_values(['amount_spent'],ascending=False)
amount_spent
```

```python
Q95 = amount_spent['amount_spent'].quantile(0.95)
Q75 = amount_spent['amount_spent'].quantile(0.75)

Q95
Q75
```

```python
def customer_group(x):
    if x['amount_spent'] >= Q95:
        return 'VIP'
    elif x['amount_spent'] >= Q75 and x['amount_spent'] < Q95:
        return 'Preferred'
    return 'Regular'
```

```python
amount_spent['Group'] = amount_spent.apply(lambda x: customer_group(x), axis=1)
customers_group = amount_spent[['CustomerID','Group']]
customers_group
```

## Q2: How to identify which country has the most VIP Customers?

```python
df1=pd.merge(df,amount_spent, on='CustomerID')

```

```python
columns = ['CustomerID','Country','Group']
country_customer = pd.DataFrame(df1.groupby(columns).groups.keys(), columns=columns)
country_customer 
```

```python
vips = country_customer[country_customer['Group'] == 'VIP']
vips
```

```python
cct = vips[['Country', 'CustomerID']].groupby('Country').count().nlargest(1,'CustomerID')
cct
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
vip_pref = country_customer[country_customer['Group'].isin(['VIP','Preferred'])]
```
